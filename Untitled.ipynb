```python
df = pd.get_dummies(df,columns =['explicit','mode'])
```

```python
del df['explicit_No']
```

```python
del df['mode_Minor']
```

```python
import datetime as dt
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_date']= df['release_date'].map(dt.datetime.toordinal)
```

```python
df.corr()
```

```python
sns.jointplot(x=df['energy'],y=(df['acousticness']))
```

```python
x_multi = df.drop(['popularity'],axis=1)
```

```python
y_multi = df['popularity']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_multi,y_multi,test_size = 0.1,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 1500,n_jobs = -1,random_state = 42)
```

```python
rfc.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,rfc.predict(x_test))
```

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
scaler = QuantileTransformer()
```

```python
x_trains = scaler.fit_transform(x_train)
```

```python
x_tests = scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion = 'gini',max_depth = 5,min_samples_split= 1000,random_state = 42)
```

```python
dtc.fit(x_train,y_train)
```

```python
accuracy_score(y_test,dtc.predict(x_test))
```

```python
rfc2 = RandomForestClassifier(n_estimators = 1500,min_samples_split = 30,bootstrap= False,n_jobs = -1,random_state = 42)
```

```python
rfc2.fit(x_train,y_train)
```

```python
accuracy_score(y_test,rfc2.predict(x_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(rfc,learning_rate = 0.02,n_estimators = 1500)
```

```python
ada.fit(x_train,y_train)
```

```python
accuracy_score(y_test,ada.predict(x_test))
```

```python
from xgboost import XGBRFClassifier
```

```python
xgb = XGBRFClassifier(n_estimators=1000)
```

```python
xgb.fit(x_train,y_train)
```

```python
accuracy_score(y_test,xgb.predict(x_test))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'rbf',random_state = 42)
```

```python
svc.fit(x_trains,y_train)
```

```python
accuracy_score(y_test,svc.predict(x_tests))
```
